```python
import json
import requests
import os
import pandas as pd
import numpy as np
from pprint import pprint
from config import api_key
import matplotlib.pyplot as plt
```

```python
#Take in all of our OES data from 2014 to 2018 and read it into pandas
OES_2014_df = pd.read_csv('Resources/state_M2014_dl.csv')
OES_2015_df = pd.read_csv('Resources/state_M2015_dl.csv')
OES_2016_df = pd.read_csv('Resources/state_M2016_dl.csv')
OES_2017_df = pd.read_csv('Resources/state_M2017_dl.csv')
OES_2018_df = pd.read_csv('Resources/state_M2018_dl.csv')
```

```python
print(len(OES_2014_df))
print(len(OES_2015_df))
print(len(OES_2016_df))
print(len(OES_2017_df))
print(len(OES_2018_df))
```

```python
#Extract the columns we want
OES_2014_df = OES_2014_df.loc[:, ["AREA", "STATE", "OCC_CODE","OCC_TITLE", "TOT_EMP"]]
OES_2015_df = OES_2015_df.loc[:, ["AREA", "STATE", "OCC_CODE","OCC_TITLE", "TOT_EMP"]]
OES_2016_df = OES_2016_df.loc[:, ["AREA", "STATE", "OCC_CODE","OCC_TITLE", "TOT_EMP"]]
OES_2017_df = OES_2017_df.loc[:, ["AREA", "STATE", "OCC_CODE","OCC_TITLE", "TOT_EMP"]]
OES_2018_df = OES_2018_df.loc[:, ["AREA", "STATE", "OCC_CODE","OCC_TITLE", "TOT_EMP"]]
```

```python
# Merge the first two datasets on "OCC_CODE"
combined_OES_df = pd.merge(OES_2014_df, OES_2015_df,
                                 how='outer', on=('STATE','OCC_CODE'))
```

```python
#Rename columns
combined_OES_df = combined_OES_df.rename(columns={"AREA_x":"2014 AREA",
                                                              "OCC_TITLE_x":"2014 OCC TITLE",
                                                              "TOT_EMP_x":"2014 TOT EMP"})

combined_OES_df = combined_OES_df.rename(columns={"AREA_y":"2015 AREA",
                                                              "OCC_TITLE_y":"2015 OCC TITLE",
                                                              "TOT_EMP_y":"2015 TOT EMP"})
combined_OES_df.head()
```

```python
# Merge our newly combined dataframe with the 2016 dataframe
combined_OES_df = pd.merge(combined_OES_df, OES_2016_df, how="outer", on=('STATE','OCC_CODE'))
combined_OES_df.head()
```

```python
#Rename columns
combined_OES_df = combined_OES_df.rename(columns={"AREA":"2016 AREA",
                                                              "OCC_TITLE":"2016 OCC TITLE",
                                                              "TOT_EMP":"2016 TOT EMP"})
combined_OES_df.head()
```

```python
# Merge our newly combined dataframe with the 2017 dataframe
combined_OES_df = pd.merge(combined_OES_df, OES_2017_df, how="outer", on=('STATE','OCC_CODE'))
combined_OES_df.head()
```

```python
#Rename columns
combined_OES_df = combined_OES_df.rename(columns={"AREA":"2017 AREA",
                                                              "OCC_TITLE":"2017 OCC TITLE",
                                                              "TOT_EMP":"2017 TOT EMP"})
combined_OES_df.head()
```

```python
# Merge our newly combined dataframe with the 2017 dataframe
combined_OES_df = pd.merge(combined_OES_df, OES_2018_df, how="outer", on=('STATE','OCC_CODE'))
combined_OES_df.head()
```

```python
#Rename columns
combined_OES_df = combined_OES_df.rename(columns={"AREA":"2018 AREA",
                                                              "OCC_TITLE":"2018 OCC TITLE",
                                                              "TOT_EMP":"2018 TOT EMP"})
combined_OES_df.head()
```

```python
#Select the columns we want
combined_OES_df = combined_OES_df.loc[:, ["2014 AREA", "STATE", "OCC_CODE",
                                          "2014 OCC TITLE", "2014 TOT EMP",
                                          "2015 OCC TITLE", "2015 TOT EMP",
                                         "2016 OCC TITLE", "2016 TOT EMP",
                                         "2017 OCC TITLE", "2017 TOT EMP",
                                         "2018 OCC TITLE", "2018 TOT EMP"]]
combined_OES_df.count()
```

```python
#filter out the data, leaving with the data of our selected occuaptions
filter1 = combined_OES_df['OCC_CODE'].isin(['13-1161','13-1151','13-1041','13-1111','13-1051','15-1132',
                                           '15-1141','15-1122','15-1121','15-1131','17-2141','17-2051',
                                            '17-2061','17-2071','17-2011'])
clean_OES_df = combined_OES_df[filter1] 
```

```python
#delete all rows which are not from the 50 states
clean_OES_df = clean_OES_df[clean_OES_df['STATE'].isin(['District of Columbia','Puerto Rico','Virgin Islands','Guam']) == False]
clean_OES_df.head()
```

```python
#Select the columns we want
clean_OES_df = clean_OES_df.loc[:, ["STATE", "OCC_CODE","2014 OCC TITLE", "2014 TOT EMP",
                                   "2015 TOT EMP","2016 TOT EMP","2017 TOT EMP","2018 TOT EMP"]]
clean_OES_df.head()
```

```python
#Rename column
clean_OES_df = clean_OES_df.rename(columns={"2014 OCC TITLE":"OCC TITLE"})
clean_OES_df.head()
```

```python
#Convert it into a dataFrame
df_clean_OES = pd.DataFrame(clean_OES_df)
```

```python
#remove rows with Null/NaN values for each year
df_clean_OES.dropna(subset=['2014 TOT EMP'],inplace=True)
df_clean_OES.dropna(subset=['2015 TOT EMP'],inplace=True)
df_clean_OES.dropna(subset=['2016 TOT EMP'],inplace=True)
df_clean_OES.dropna(subset=['2017 TOT EMP'],inplace=True)
df_clean_OES.dropna(subset=['2018 TOT EMP'],inplace=True)
```

```python
#Replace ** with 0 so that we can perform summation later
df_clean_OES.replace({"**": "0"}, inplace=True)
df_clean_OES
```

```python
#Seperate the 15 occupations into individual dataframes
marketingSpecialists = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '13-1161']
trainingSpecialists = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '13-1151']
complianceOfficers = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '13-1041']
managementAnalysts = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '13-1111']
costEstimator = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '13-1051']
softwareDevelopers = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '15-1132']
databaseAdministrators = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '15-1141']
informationAnalysts = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '15-1122']
computerSystemsAnalysts = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '15-1121']
computerProgrammers = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '15-1131']
mechanicalEngineers = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '17-2141']
civilEngineers = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '17-2051']
computerHardwareEngineers = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '17-2061']
electricalEngineers = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '17-2071']
aerospaceEngineers = df_clean_OES.loc[df_clean_OES['OCC_CODE'] == '17-2011']
```
