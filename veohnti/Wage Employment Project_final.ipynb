# Computer Occupations

```python

#Creating my open canvas
years = [2014,2015,2016,2017,2018]

fig, ax = plt.subplots()
ax.plot(years, soft_dev_wages, alpha = 0.8, marker = 'o', label = "Software Developer", linewidth=3.0, markersize =10.0)
ax.plot(years, data_admin_wages, alpha = 0.8, marker = '^', label = "Database Administrator", linewidth=3.0, markersize =10.0)
ax.plot(years, info_sec_analyst_wages, alpha = 0.8, marker = '*', label="Information Security Analyst", linewidth=3.0, markersize =10.0)
ax.plot(years, sys_analyst_wages, alpha = 0.8, marker = '<', label = "Computer System Analyst", linewidth=3.0, markersize =10.0)
ax.plot(years, comp_programmer_wages, alpha = 0.8, marker = '<', label = "Computer Programmer", linewidth=3.0, markersize =10.0)

fig;

# Creating my grid and labels

ax.set(xlabel = "Years", ylabel="Average Hourly Wage ($)", title="Computer Occupations Wage Change");
ax.grid('on', which='major');
plt.xticks(years);
ax.legend(loc="best")
ax.set_ylim(20, 65);


# Saving the file
path = os.path.join(os.path.expanduser("~"), "Desktop", "Computer Occupations Wage Change.png")
path
fig.savefig(path)

# Making it bigger

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(10, 7)
```

Observation:
In this graph, we see that wages of 'Computer Occupations' have positively changed in a similar manner. We can especially see that Computer Programmers and Database Administrators have very similar salaries.
However, if you are looking for a higher paid job, this graph suggest that you should try to become a Software Developer to have a higher salary
