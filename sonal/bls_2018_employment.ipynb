```python
# Now we need to begin to build our data dictionary. Easiest way to do this is to use the dict() function of the general form:
# type = 'choropleth',
# locations = list of states
# locationmode = 'USA-states'
# colorscale= 
# Either a predefined string:
# 'pairs' | 'Greys' | 'Greens' | 'Bluered' | 'Hot' | 'Picnic' | 'Portland' | 'Jet' | 'RdBu' | 'Blackbody' | 'Earth' | 'Electric' | 'YIOrRd' | 'YIGnBu'
```

```python
bls3['text'] = bls3['STATE_NAME'] + '<br>' + \
    'JOBS_PER_1000 :' + bls3['JOBS_PER_1000'] +'<br>' + \
    ' LOCATION_QUOTIENT : ' + bls3['LOCATION_QUOTIENT'] + '<br>' + \
    'ANNUAL_MEAN_WAGE :' + bls3['ANNUAL_MEAN_WAGE']
bls3
```

```python
data = dict(type='choropleth',
            colorscale = 'Portland',
            locations = bls3['STATE_CODE'],
            z = bls3['TOTAL_EMPLOYMENT'],
            locationmode = 'USA-states',
            text = bls3['text'],
            marker = dict(line = dict(color = 'rgb(255,255,255)',width = 2)),
            colorbar = {'title':"TOTAL_EMPLOYMENT"}
            )
```

```python
layout = dict(title = '2018 US Employment by State',
              geo = dict(scope='usa'))
```

```python
choromap = go.Figure(data = [data],layout = layout)
```
